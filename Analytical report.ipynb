```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
from datetime import date
from dateutil.relativedelta import relativedelta
from datetime import datetime, timedelta
import dateutil.relativedelta

course = pd.read_csv('courses.csv', sep=',') #загружаем courses.csv в основной датафрейм
course.rename(columns={'id':'course_id'}, inplace=True) #переименовываем колонку для дальнейшего объединения
course_c = pd.read_csv('course_contents.csv', sep=',') #загружаем course_contents.csv
prog = pd.read_csv('progresses.csv', sep=',') #загружаем progresses.csv
prog.rename(columns={'id':'progress_id'}, inplace=True) #переименовываем колонку для дальнейшего объединения
prog_p = pd.read_csv('progress_phases.csv', sep=',', parse_dates=['start_date', 'finish_date'], dayfirst=True) #загружаем
# progress_phases.csv
main = prog_p # присваиваем основному датафрейму загруженны файл
prog = prog.merge(course.iloc[:, 1:4], how='left') # объединяем датафреймы по признаку course_id
prog = prog.merge(course_c, how='left') # объединяем датафреймы по признаку course_id
main = main.merge(prog, how='left') # объединяем датафреймы по признаку progress_id
main[main.student_id == '768c2987a744c51ce64a5993a2a94eaf'].head() # проверка основного датасета
```

```python
course_sum = main.groupby('title')['lesson_number'].count().reset_index() # собираем все типы курсов
course_sum = course_sum.assign(dummy = 1) # добавляем столбец для счета
print("Всего курсов", course_sum['dummy'].sum()) # выводим кол-во курсов
```

```python
modul_sum = main.groupby(['title','module_number'])['lesson_number'].count().reset_index() # собираем все модули
modul_sum = modul_sum.assign(dummy_1 = 1) # добавляем столбец для счета
modul_sum = modul_sum.groupby('title')['dummy_1'].count().reset_index() # считаем кол-во модулей в каждом курсе
modul_sum.sort_values(['dummy_1'], ascending=False, inplace=False) # выводим результат
```

```python
lesson_sum = main.groupby(['title','module_title', 'lesson_number'])['module_number'].count().reset_index() # собираем все уроки
lesson_sum = lesson_sum.assign(lessons_sum = 1) # добавляем столбец для счета
# считаем кол-во уроков в каждом модуле и курсе
lesson_sum = lesson_sum.groupby(['title','module_title'])['lessons_sum'].count().reset_index() 
# считаем кол-во уроков на курсе
lesson_sum_1 = lesson_sum.groupby('title')['lessons_sum'].sum().reset_index() 
lesson_sum_1.sort_values(['lessons_sum'], ascending=False, inplace=False) # выводим результат
```

```python
# считаем медианное значение кол-ва уроков на каждый модуль
lesson_sum_2 = lesson_sum.groupby('title')['lessons_sum'].median().reset_index()
lesson_sum_2.sort_values(['lessons_sum'], ascending=False, inplace=False) # выводим результат
```

```python
student_sum = main.groupby(['title', 'field', 'student_id'])['module_number'].count().reset_index() # собираем всех студентов
student_sum = student_sum.assign(students_sum = 1) # добавляем столбец для счета
# считаем кол-во студентов на каждом курсе
student_sum = student_sum.groupby(['title', 'field'])['students_sum'].count().reset_index() 
student_sum.sort_values(['students_sum'], ascending=False, inplace=False) # выводим результат
```

```python
stud = pd.read_csv('students.csv', sep=',', parse_dates=['birthday'], dayfirst=True) #загружаем students.csv
stud.rename(columns={'id':'student_id'}, inplace=True) #переименовываем колонку для дальнейшего объединения
# делаем отедльный датафрем из объединения с другими файлами
stud = stud.merge(prog, how='left')
stud = stud.merge(course, how='left')
stud = stud.dropna() # удаляем пустые значения
now_day = date.today() # создаем столбец с текущей датой
stud['birthday'] = pd.to_datetime(stud['birthday'], errors = 'coerce') # переводим столбец дней рождений в формат datetime
# создаем функцию расчета возраста
def age(row):   
    birthday = row['birthday']
    if pd.isnull(birthday) or pd.isnull(now_day): 
        return 0
    return relativedelta(now_day, birthday).years
stud['age'] = stud.apply(age, axis=1) # применяем функцию к столбцу с возрастом
stud = stud[stud['age'] > 18] # ставим такую рамку, потому что в процессе были выявленны дни рождения за текущей датой
stud_min = stud['age'].min() # Находим минимальный возраст студентов
stud_mean = stud['age'].mean() # Находим средний возраст студентов
stud_max = stud['age'].max() # Находим максимальный возраст студентов
stud_median = stud['age'].median() # Находим медианный возраст студентов
print('Минимальный возраст студентов:', stud_min, 'Средний возраст студентов:', stud_mean,
     'Максимальный возраст студентов:', stud_max, 'Медианный возраст студентов:', stud_median)
```

```python
stud_field_min = stud.groupby('title')['age'].min().reset_index() # Находим минимальный возраст студентов на курсах
stud_field_mean = stud.groupby('title')['age'].mean().reset_index() # Находим средний возраст студентов на курсах
stud_field_max = stud.groupby('title')['age'].max().reset_index() # Находим максимальный возраст студентов на курсах
stud_field_median = stud.groupby('title')['age'].median().reset_index() # Находим медиальный возраст студентов на курсах
# Выводим результат
print(stud_field_min) 
print(stud_field_mean) 
print(stud_field_max) 
print(stud_field_median)
```

```python
# Создаем датасеты с требованиями задания по датам и ДЗ. Создаем столбец для счета и осортировываем даты
group_stud = main.copy()
group_stud['start_date'] = pd.to_datetime(group_stud['start_date']).dt.strftime('%Y-%m')
group_stud['count'] = 1
group_lm = group_stud.copy()
# Выбираем домашние работы
group_lm = group_lm[(group_lm['start_date'] >= "2016-03") & (group_lm['start_date'] <= '2019-06') &
                     (group_lm['is_homework'] == True)] 
group_lm = group_lm.groupby(['title', 'start_date', 'module_number'])['count'].count().reset_index()
# Создаем отдельные датафреймы, чтобы выбрать только первые дз
k = stud_field_min['title'].tolist()
group_min_hm = pd.DataFrame()
for i in k:
    group_lim = group_lm[group_lm['title'] == i]
    j = group_lim['module_number'].min()
    group_lim = group_lim[group_lim['module_number'] == j]
    group_min_hm = pd.concat([group_min_hm, group_lim], ignore_index=True)
# Группипуем и создаем сводную таблицу с результатом прироста
pivoted = pd.pivot_table(group_min_hm, index='title', columns=['start_date'], values='count')
pivoted.head()
```

Как вариант метрики можно рассмотреть конверсию неуспеваймости по домашнему заданию - отношению невыполненого и начатого дз к выполненому. Так можно определить на каких курсах больше всего неуспевающих студентов и предпринять меры для данных курсов

```python
group_5 = main.copy()
group_5 = group_5[(group_5['lesson_title'] == 'Домашняя работа')] 
group_5['count'] = 1
# Считаем кол-во домашних заданий на модуле
group_5_fail = group_5[(group_5['status'] != 'done')].groupby(
    ['title', 'module_number', 'student_id'])['count'].count().reset_index()
group_5_fail['count'] = 1
group_5_fail = group_5_fail.dropna()
group_5_fail = group_5_fail.groupby(['title', 'module_number'])['count'].sum().reset_index()
group_5_done = group_5[(group_5['status'] == 'done')].groupby(
    ['title', 'module_number', 'student_id'])['count'].count().reset_index()
group_5_done['count'] = 1
group_5_done = group_5_done.dropna()
group_5_done = group_5_done.groupby(['title', 'module_number'])['count'].sum().reset_index()
group_5_sum = group_5_fail.copy()
group_5_sum['count_done'] = group_5_done['count']
# Cоздаем сводную таблицу
pivoted_5_fl = pd.pivot_table(group_5_sum, index='title', columns=['module_number'], values='count')
pivoted_5_dn = pd.pivot_table(group_5_sum, index='title', columns=['module_number'], values='count_done')
# Создаем дополнительный датафрейм для расчета конверсии
pivoted_5_lim = pd.DataFrame(index=pivoted_5_fl.index)
# Добавляем конверсию
i = 1
k = pivoted_5_fl.columns.values.tolist()
while i < len(k):
    pivoted_5_lim['Конверсия невыполнения' + ' ' + str(k[i-1]) + '-' + str(k[i])] = (pivoted_5_fl[i + 1] / pivoted_5_dn[i])
    i += 1
# Выводим результат
pivoted_5_lim.head()

# Делаем датафрейм для графика
pivot_5_graph = pivoted_5_lim.T.sum()
# Готовим поля для графиков
fig,ax = plt.subplots(figsize=(12,6))
# Строим графики
ax.bar(pivoted_5_lim.index, pivot_5_graph)
# формируем обозначения графика
ax.set_title('Конверсия невыполнения домашнего задания')
ax.set_xlabel('Курсы')
ax.set_ylabel('Конверсия')
ax.tick_params(axis='x', rotation=90)
# выводим результат
plt.show()
```

```python
# Создаем датасеты для определения времения выполнения дз. Создаем столбец для счета и осортировываем даты
group_hm_5 = main.copy()
group_hm_5['finish_date'] = pd.to_datetime(group_hm['finish_date'])
group_hm_5['start_date'] = pd.to_datetime(group_hm['start_date'])
group_hm_5 = group_hm_5[(group_hm_5['finish_date'] >= "2016-03") & (group_hm_5['finish_date'] <= '2019-06') &
                     (group_hm_5['lesson_title'] == 'Домашняя работа')]  
group_hm_5['count'] = 1
# Приводим неизвестные дз к дате конца выборки, чтобы учесть сроки выполнения дз
fn_dat_5 = group_hm_5['finish_date'].tolist()
for i in fn_dat_5:
    if i == 'NaN':
        j = fn_dat_5.index(m)
        fn_dat_5.pop(j)
        fn_dat_5.insert(j, '2019-06')
group_hm_5['finish_date'] = fn_dat_5
# Группипуем и создаем сводную таблицу
group_hm_lm_5 = group_hm_5.groupby(['title', 'module_number', 'student_id', 'start_date',
                                'finish_date'])['count'].count().reset_index()
# Создаем разницу дат для учета переходных месяцев
group_hm_lm_5['start_date_y'] = group_hm_lm_5['start_date'].dt.year
group_hm_lm_5['start_date_m'] = group_hm_lm_5['start_date'].dt.month
group_hm_lm_5['finish_date_y'] = group_hm_lm_5['finish_date'].dt.year
group_hm_lm_5['finish_date_m'] = group_hm_lm_5['finish_date'].dt.month
group_hm_lm_5['Разница дат'] = (group_hm_lm_5['finish_date_y'] - group_hm_lm_5['start_date_y']) * 12 + (
    group_hm_lm_5['finish_date_m'] - group_hm_lm_5['start_date_m'])
group_hm_lm_5['count'] = 1 + group_hm_lm_5['Разница дат'] # учитываем выполнение дз в несколько месяцев
group_hm_lm_5['finish_date'] = pd.to_datetime(group_hm_lm_5['finish_date']).dt.strftime('%Y-%m')
group_hm_lm_5['start_date'] = pd.to_datetime(group_hm_lm_5['start_date']).dt.strftime('%Y-%m')
group_hm_lm_5 = group_hm_lm_5.groupby(['title', 'student_id'])['count'].sum().reset_index()
mean_hm_time = group_hm_lm_5['count'].mean() #Считаем среднее время выполнения дз
group_hm_lm_5['Отношение к среднему'] = group_hm_lm_5['count'] / mean_hm_time
# Выаодим результат
group_hm_lm_5.head()
```
